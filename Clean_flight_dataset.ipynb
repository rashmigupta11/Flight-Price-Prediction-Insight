```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey',axis=1, inplace=True)
df
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
df
```

```python

df.drop('Arrival_Time',axis=1, inplace= True)
df
```

```python
df['Arrival_Hour']= df['Arrival_Hour'].astype(int)
df['Arrival_Minute']= df['Arrival_Minute'].astype(int)
df.info()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Min'] = df['Dep_Time'].str.split(':').str[1]
df
```

```python
df.drop('Dep_Time', axis=1, inplace= True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OHE= OneHotEncoder()
```

```python
cat_num_arr=OHE.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()
cat_num_arr
```

```python
cleaned_df= pd.DataFrame(cat_num_arr, columns= OHE.get_feature_names_out())
cleaned_df
```

```python
df.drop('Airline',axis= 1, inplace= True)
df.drop('Source',axis= 1, inplace= True)
df.drop('Destination',axis= 1, inplace= True)
df


```

```python
# Reset index just to be sure both are aligned
df.reset_index(drop=True, inplace=True)
cleaned_df.reset_index(drop=True, inplace=True)

# Merge the two DataFrames column-wise
final_df = pd.concat([df, cleaned_df], axis=1)

# Check the shape and preview
print(final_df.shape)
final_df.head()
```

```python
final_df.to_csv('Cleaned_flight_dataset.csv', index=False)
```
